```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('Sales_Lead_Dataset.xlsx')
df['Lead Create Date'] = pd.to_datetime(df['Lead Create Date'])
```

```python
# All time
print("=" * 55)
print("KPI CARDS")
print("=" * 55)
total = len(df)
enrolled = df[df['Lead Status'] == 'Enrolled']
no_resp = df[df['Lead Status'] == 'No Response']

print(f"Total Leads:                {total}")
print(f"Enrolled:                   {len(enrolled)}")
print(f"Enrollment Rate:            {round(len(enrolled)/total*100, 1)}%")
print(f"No Response:                {len(no_resp)}")
print(f"No-Response Rate:           {round(len(no_resp)/total*100, 1)}%")
print(f"Avg Follow-Ups (Enrolled):  {round(enrolled['Follow Up Count'].mean(), 2)}")
print(f"Avg Follow-Ups (No Resp):   {round(no_resp['Follow Up Count'].mean(), 2)}")

print("\n" + "=" * 55)
print("INSIGHT STRIP")
print("=" * 55)
grp_yes = df[df['Group Connected'] == 'Yes']
grp_no  = df[df['Group Connected'] == 'No']
rate_yes = round(len(grp_yes[grp_yes['Lead Status']=='Enrolled']) / len(grp_yes) * 100, 1)
rate_no  = round(len(grp_no[grp_no['Lead Status']=='Enrolled'])  / len(grp_no)  * 100, 1)
print(f"Group connection advantage: +{round(rate_yes - rate_no)}pt  ({rate_yes}% vs {rate_no}%)")
print(f"No-response leads:          {len(no_resp)}")
print(f"Avg follow-ups (No Resp):   {round(no_resp['Follow Up Count'].mean(), 1)}")

df['Lead Create Date'] = pd.to_datetime(df['Lead Create Date'])
df['Month'] = df['Lead Create Date'].dt.to_period('M')
monthly_vol = df.groupby('Month').size()
peak_month  = monthly_vol.idxmax()
trough_month = monthly_vol.idxmin()
print(f"Peak month:                 {peak_month} ({monthly_vol[peak_month]} leads)")
print(f"Trough month:               {trough_month} ({monthly_vol[trough_month]} leads)")
print(f"Peak vs trough ratio:       {round(monthly_vol[peak_month] / monthly_vol[trough_month], 1)}x")

print("\n" + "=" * 55)
print("MONTHLY LEAD VOLUME")
print("=" * 55)
print(monthly_vol.to_string())

print("\n" + "=" * 55)
print("LEAD STATUS BREAKDOWN")
print("=" * 55)
status_counts = df['Lead Status'].value_counts()
print(status_counts.to_string())
other = status_counts.drop(['Enrolled','No Response','Connecting'], errors='ignore').sum()
print(f"Other (Lost+Expired+Canceled): {other}")

print("\n" + "=" * 55)
print("ENROLLMENT RATE BY STUDENT CATEGORY")
print("=" * 55)
for cat, grp in df.groupby('Student Category'):
    rate = round(len(grp[grp['Lead Status']=='Enrolled']) / len(grp) * 100, 1)
    print(f"  {cat:<15} {len(grp):>3} leads   {rate}% enrolled")

print("\n" + "=" * 55)
print("GROUP CONNECTION IMPACT")
print("=" * 55)
for gc, grp in df.groupby('Group Connected'):
    e = len(grp[grp['Lead Status']=='Enrolled'])
    ne = len(grp) - e
    rate = round(e / len(grp) * 100, 1)
    print(f"  {gc:<5}  total={len(grp)}  enrolled={e}  not enrolled={ne}  rate={rate}%")

print("\n" + "=" * 55)
print("POLICY DISTRIBUTION (enrolled only)")
print("=" * 55)
policy_counts = enrolled['Latest Policy'].value_counts()
print(policy_counts.to_string())

print("\n" + "=" * 55)
print("SALES AGENT PERFORMANCE")
print("=" * 55)
for agent, grp in df.groupby('Sales Agent'):
    e = len(grp[grp['Lead Status']=='Enrolled'])
    ne = len(grp) - e
    rate = round(e / len(grp) * 100, 1)
    print(f"  {agent:<12} total={len(grp)}  enrolled={e}  not enrolled={ne}  rate={rate}%")

print("\n" + "=" * 55)
print("AVG FOLLOW-UPS BY OUTCOME")
print("=" * 55)
fu_by_status = df.groupby('Lead Status')['Follow Up Count'].mean().round(2).sort_values()
print(fu_by_status.to_string())
```
